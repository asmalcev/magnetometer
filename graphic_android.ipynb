```python
import numpy as np
import csv
import matplotlib.pyplot as plt
from functools import reduce
```

```python
data = []

with open('dist/o.csv', 'r') as fr:
  csvreader = csv.reader(fr, delimiter=';')

  for row in csvreader:
    data.append(list(map(float, row)))
```

```python
def pass_elems_to_tuple(d, i1, i2, tuple_arrays):
  tuple_arrays[0].append(d[i1])
  tuple_arrays[1].append(d[i2])
  return tuple_arrays

plt.ylabel('Смещение (h), µT')
plt.xlabel('Время, с')

(x, y) = reduce(lambda acc, current: pass_elems_to_tuple(current, 0, 1, acc), data, ([], []))
plt.plot(x, y, label='x')

(x, y) = reduce(lambda acc, current: pass_elems_to_tuple(current, 0, 2, acc), data, ([], []))
plt.plot(x, y, label='y')

(x, y) = reduce(lambda acc, current: pass_elems_to_tuple(current, 0, 3, acc), data, ([], []))
plt.plot(x, y, label='z')

plt.legend()
plt.show()
```

```python
labels = ['d11', 'd22', 'd33', 'd12', 'd13', 'd23']
for i in range(4, 10):
  (x, y) = reduce(lambda acc, current: pass_elems_to_tuple(current, 0, i, acc), data, ([], []))
  plt.xlabel('Время, с')
  plt.plot(x, y, label=labels[i - 4])
  plt.legend()
  plt.show()
```
